# 1. How can we count the words in a text?


```python
# Example sentence
chomsky = 'colourless green ideas sleep furiously'
print(chomsky)
```

   


---



 

---



 

---





---



```python
# Split a sentence into words
import nltk
chomsky_words = nltk.word_tokenize(chomsky)

print(chomsky_words)
```

   


---



 

---



 

---





---



```python
# measure word frequency
frequency = nltk.FreqDist(chomsky_words)
frequency
```

   


---



 

---



 

---





---



```python
# a more complex example
turtles = """teenage mutant ninja turtles, 
            teenage mutant ninja turtles, 
            teenage mutate ninja turtles, 
            heroes in a halfshell, turtle power!"""

print(turtles)
```

   


---



 

---



 

---





---



```python
turtle_words = nltk.word_tokenize(turtles)
print(*turtle_words, sep = '\n')
```

   


---



 

---



 

---





---



```python
# create a frequency distribution
turtle_frequency = nltk.FreqDist(turtle_words)

# query a specific word
print(turtle_frequency['ninja'])
```

# 2. How diverse is the vocabulary?

```python
# Create simple fuction to measure diversity
def lexical_diversity(input):
  # divide the number of unique words by the number of total words
  return len(set(input))/len(input)
```

   


---



 

---



 

---





---



```python
# Every word in the Chomsky quote is unique
lexical_diversity(chomsky_words)
```

   


---



 

---



 

---





---



```python
# The TMNT theme song has repeated words (less diverse)
lexical_diversity(turtle_words)
```

# 3. What is the emotional valence of a text?

```python
# Import NLTK sentiment analyzer
from nltk.sentiment import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
```

   


---



 

---



 

---





---



```python
# calculate sentiment of various examples
sid.polarity_scores('I hate pizza!')['compound']
```

   


---



 

---



 

---





---



```python
sid.polarity_scores('I love pizza!')['compound']
```

   


---



 

---



 

---





---



```python
sid.polarity_scores('I have a love hate relationship with pizza!')['compound']
```
