# **What is a Colab Notebook?**




##**Step 1: Create a new Colab notebook**

Make sure you are [signed in to Google Drive](https://drive.google.com). If you have not yet created a Google Drive, please see [this link](https://edu.gcfglobal.org/en/googledriveanddocs/getting-started-with-google-drive/1/) for how to get started with Google Drive.


Once you are signed in, click on [this link to create a new Colab notebook](http://colab.research.google.com/#create=true).


You will see something like this:
>![New Colab](https://drive.google.com/uc?export=view&id=1Lh-m2zaNy4fCTa5Nr8NwsgISGVhbGxJK)






## **Step 2: Give your notebook a title**
Click on the file name:
>![Untitled](https://drive.google.com/uc?export=view&id=1L2AX13NFXJEiJNDILEuh-gNkATsWJNpo)

Change the file name to something more memorable:
>![Rename](https://drive.google.com/uc?export=view&id=1_ira4XiqTCo6GgWtc7HDXwGvEDI-lZIJ)



## **Step 3: Locate your Colab in your Google Drive**

If this is your first time creating a Google Colab notebook, Google should automatically create a new (orange!) folder in your Google Drive. [Go to your Google Drive](https://drive.google.com) and see if you can find this folder in the sidebar:

><img src = "https://drive.google.com/uc?export=view&id=1nSG8IMBUt7CbsxIw6njfQ3IrYU0LFN1M" height = 200>

Click on this folder and you should see your new notebook:

><img src = "https://drive.google.com/uc?export=view&id=1R7QpG_h4YtMY_q0hKn7lkmWgh88K-5ha" height = 300>

This is now your default folder for your Colab notebooks. Whenever you create a new notebook, or make a copy of another notebook, it will automatically be saved in this folder.

You should now also be able to create a new Colab notebook directly from your Drive, in whichever folder you like, without needing to follow the link in Step 1. To do so, click on the "New" button on the top of your sidebar, click "More" and select "Google Colaboratory": 

><img src = "https://drive.google.com/uc?export=view&id=1dT0Y0mhALn9S1z1q0lD-iOpDkKqhWYAh" height = 300>

If you don't see this option, select "Connect more apps" and search for  Colaboratory:

>![Connect Apps](https://drive.google.com/uc?export=view&id=1CiKrRO81cbMaoIipZyMr_85STYxMwvLE)

## **Step 4: Delete the default Code cell**

When you create a new notebook, Colab will automatically include a blank Code cell, which looks like this:
>![Code Cell](https://drive.google.com/uc?export=view&id=1cnPxPw4YHEMcV3K8Jj735aSBNZYrbi49)

As we are only working with text at the moment, you should delete this by clicking on the rubbish bin on the far right of the cell:
>![Delete Cell](https://drive.google.com/uc?export=view&id=1CjcA4Oh2dtRg_ggAONe1wKyKTzuxE1L0)


## **Step 5: Create a Text cell**

Click on the **+ Text** button to add a new Text cell:
>![Add Text Cell](https://drive.google.com/uc?export=view&id=1WT4bKPDiYuKKY6FAlJPz4PoPdH_GDvsI)

Your blank text cell will look like this:
>![Blank Text Cell](https://drive.google.com/uc?export=view&id=1yFT9yAryAOPTEwA456q6VFzFlxUhb8P3)

You will notice that your cell is divided into two columns. The left hand column is where you can insert text, links, images, and more, using a language called **Markdown**. The right-hand column shows what this looks like to someone viewing your notebook.

Try typing something into the left-hand column. You will notice the same text appear simulatenously in the right-hand column:
>![Sample markdown](https://drive.google.com/uc?export=view&id=14TxP5fmpyzmGovmG39Unrepdb5IUkJp8)

Double-click on the right-hand column to just see the displayed output:
>![Sample output](https://drive.google.com/uc?export=view&id=1ZllmatKTfRyiq9ctl6D07HwfqgPLe2R4)

You can then double-click on the cell again to edit it some more. 

If you want to add another cell, you can click on **+ Text** again to create a new cell underneath the previous one.

## **Step 6: Play around and have some fun**

Now's probably a good time to play around with Markdown and see if you can get the hang of it. 

You can have a look at Google's guide to [Getting Started with Colab](https://colab.research.google.com/notebooks/intro.ipynb#scrollTo=GJBs_flRovLc), [Overview of Colaboratory Features](https://colab.research.google.com/notebooks/basic_features_overview.ipynb) and [Overview of Markdown](https://colab.research.google.com/notebooks/markdown_guide.ipynb#scrollTo=Lhfnlq1Surtk) to help you.

Experiment with bolding and italicising text, resizing text, adding links, and so on. 

Things can start to get a little complicated when you try to insert images. Click [here](https://colab.research.google.com/drive/1auxhazYisj6Ze2lhsnONcn9PQMfkhic8) for a tutorial on inserting images.

The folder for day one also includes notebooks showing you how to share your Colab files with other people in Google Drive and how to save your own copy of Colab files that other people have created.

When you're ready, get started on the Introduce Yourself task.
